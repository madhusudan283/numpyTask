### 1.Install matplotlib and import pyplot as plt

### 2.Plot a basic line chart using plt.plot()

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.plot(x,y,color='green',linewidth=5,linestyle='-')
plt.show()
```

### 3.Add title and labels to axes

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y)
plt.show()
```

### 4.Change line color, style, and markers

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='red',marker='*',linestyle='')
plt.show()
```

### 5.Plot multiple lines in the same chart

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.show()
```

### 6.Add a legend to the chart

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")

plt.legend(loc='upper right',shadow=True,fontsize='large')
plt.show()
```

### 7.Save plot as an image file

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.savefig('Days.PNG',transparent=None,facecolor='auto',edgecolor='auto')
plt.show()
```

### 14.Plot a scatter plot using plt.scatter()

```python
x=[1,3,5,7,8]
y=[11,6,8,2,1]

plt.scatter(x,y,color='Orange',marker='H')
plt.title("Simple Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

### 15.Add annotations to plot points

### 16.Create a stacked bar chart

```python
labels=['A','B','C']
men=[20,35,30]
women=[25,32,34]
plt.bar(labels,men,label='Men')
plt.bar(labels,women,bottom=men,label='Women')
plt.xlabel("Scores")
plt.ylabel("Y Axis")
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

### 17.Change background and axis colors

```python
x=[1,2,3]
y=[10,15,10]

fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_facecolor('lightblue')
fig.patch.set_facecolor('lightyellow')
plt.show()
```

### 18.Create a plot with twin axes (twinx/twiny)

```python
x=[1,2,3,4,5]
y1=[10,20,25,30,40]
y2=[100,80,60,40,20]
fig,ax1=plt.subplots()
ax1.plot(x,y1,color='yellow')
ax1.set_ylabel('Y1 values',color='yellow')
ax2=ax1.twinx()
ax2.plot(x,y2,color='red')
ax2.set_ylabel('Y2 values',color='red')
plt.title('Twinx axes')
plt.show()
```

```python
y=[1,2,3,4,5]
x1=[10,20,25,30,40]
x2=[100,80,60,40,20]
fig,ax1=plt.subplots()
ax1.plot(x1,y,color='blue')
ax1.set_xlabel('X1 values',color='blue')
ax2=ax1.twiny()
ax2.plot(x2,y,color='pink')
ax2.set_xlabel('X2 values',color='pink')
plt.title('Twiny axes')
plt.show()
```
